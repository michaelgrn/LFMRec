```python
import scipy.sparse
sparse_matrix = scipy.sparse.load_npz('/tmp/kids_matrix.npz')
```

```python
entries = []
for entry in sparse_matrix:
    entries.append(entry)
```

```python
ratings = []
userID = 0
itemNum = 0
from tqdm import tqdm
from time import sleep
with tqdm(total=8500) as pbar:
    for sm in entries:
        itemNum = 0
        ndarray = sm.toarray()
        #listOflist = ndarray.tolist()
        for entry in ndarray:
            for entr in entry:
                #print(entr)
                if entr > 0:
                    ratings.append([userID,itemNum,entr])
                itemNum +=1 
        userID += 1
        pbar.update(1)
```

```python
import pickle
pickle.dump( ratings, open( "save.p", "wb" ) )
```

```python

from lenskit import batch, topn, util
from lenskit import crossfold as xf
from lenskit.algorithms import Recommender, als, item_knn as knn, user_knn as uknn, funksvd
from lenskit.algorithms.basic import Fallback
from lenskit import topn
from lenskit.algorithms.basic import Bias

algo_als = funksvd.FunkSVD(50)
algo_als2 = als.BiasedMF(50)

def eval(aname, algo, train, test):
    base = Bias(damping=5)
    #algo = Fallback(algo_als,algo_als2, base)
    fittable = util.clone(algo)
    fittable = Recommender.adapt(fittable)
    fittable.fit(train)
    users = test.user.unique()
    # now we run the recommender
    recs = batch.predict(fittable, test)
    # add the algorithm name for analyzability
    recs['Algorithm'] = aname
    return recs
```

```python

all_recs = []
test_data = []

for train, test in xf.partition_users(ratings[['user', 'item', 'rating']], 1, xf.SampleFrac(0.3)):
    test_data.append(test)
    #all_recs.append(eval('ItemItem', algo_ii, train, test))
    #all_recs.append(eval('UserUser',algo_uu, train, test))
    all_recs.append(eval('MF', algo_als, train, test))


```
