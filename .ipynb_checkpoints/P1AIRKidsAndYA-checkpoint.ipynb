```python
import csv

users = []
nonUsers = []
allusers = []
adultUsers = []
kids = []
kidsId = []
adultId = []
biggest = 0
with open('LFM-1b_users.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter='\t')
    line_count = -1
    for row in csv_reader:
        if line_count == -1:
            print("Column names are ",  ", ".join(row))
            line_count += 1
        elif line_count < 120175:
            allusers.append(line_count)
            if(row[1] != "" and row[2] != "-1"):
                if int(row[2]) > 5 and int(row[2]) < 26:
                    adultUsers.append(line_count)
                    adultId.append(row[0])
                    if int(row[2]) > biggest:
                        biggest = int(row[2])
                else:
                    kids.append(row)
                    kidsId.append(row[0])
            else:
                nonUsers.append(line_count)
            line_count +=1
                
```

```python
import numpy as np
import h5py
from scipy import sparse

UAM_MATLAB_FILE = 'LFM-1b_LEs.mat'                  # Matlab .mat file where the listening events are stored

def read_UAM(m_file):
    mf = h5py.File(m_file, 'r')
    #user_ids = np.array(mf.get('idx_users')).astype(np.int64)
    #artist_ids = np.array(mf.get('idx_artists')).astype(np.int64)
    # Load UAM
    UAM = sparse.csr_matrix((mf['/LEs/']["data"],
                             mf['/LEs/']["ir"],
                             mf['/LEs/']["jc"])).transpose()    #.tocoo().transpose()
    # user and artist indices to access UAM
    #UAM_user_idx = UAM.indices #UAM.row -> for COO matrix
    #UAM_artist_idx = UAM.indptr #UAM.col -> for COO matrix
    return UAM #, UAM_user_idx, UAM_artist_idx, user_ids, artist_ids
```
