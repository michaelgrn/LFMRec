```python
import scipy.sparse
sparse_matrix = scipy.sparse.load_npz('adultsF_matrix.npz')
```

```python
entries = []
for entry in sparse_matrix:
    entries.append(entry)
```

```python
ratings = []
userID = 0
itemNum = 0
from tqdm import tqdm
from time import sleep
with tqdm(total=28000) as pbar:
    for sm in entries:
        itemNum = 0
        ndarray = sm.toarray()
        #listOflist = ndarray.tolist()
        for entry in ndarray:
            for entr in entry:
                #print(entr)
                if entr > 0:
                    ratings.append([userID,itemNum,entr])
                itemNum +=1 
        userID += 1
        pbar.update(1)
```

```python
import pickle
pickle.dump( ratings, open( "saveAdultsF.p", "wb" ) )
```
